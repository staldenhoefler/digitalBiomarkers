```python
# Example 1.13: Applying a NIfTI mask
# Load image and mask
img = nib.load("./data/image_middle_slice.nii.gz").get_fdata()
mask = nib.load("./data/mask_middle_slice.nii.gz").get_fdata().astype(bool)

# Apply mask
masked_img = np.where(mask, img, np.nan)

# Display one slice
slice_idx = img.shape[2] // 2  # middle axial slice
plt.imshow(masked_img[:, :, slice_idx], cmap="gray")
plt.title("Example 10: Masked NIfTI Slice")
plt.show()

# Exercise 2.12:
# Load mask.nii.gz.
# - Select any slice.
# - Display only the masked region
# - Calculate voxel volume

mask_nii = nib.load("./data/mask.nii.gz")
mask = mask_nii.get_fdata().astype(bool)
phantom_nii = nib.load("./data/phantom.nii.gz")
img = phantom_nii.get_fdata()
masked_img = np.where(mask, img, np.nan)

slice_idx = 20
mask_slice = mask[:, :, slice_idx]

plt.imshow(masked_img[:, :, slice_idx], cmap="gray")
plt.title(f"Masked Region - Slice {slice_idx}")
plt.show()


voxel_dims = mask_nii.header.get_zooms()
voxel_volume_nibabel = np.prod(voxel_dims)
print(voxel_volume_nibabel)
```

```python
# Make animation of masked slices
import matplotlib.animation as animation
fig, ax = plt.subplots()
def update(frame):
    ax.clear()
    ax.imshow(masked_img[:, :, frame], cmap="gray")
    ax.set_title(f"Masked Slice {frame}")
ani = animation.FuncAnimation(fig, update, frames=masked_img.shape[2], interval=200)
ani.save("masked_slices.gif", writer='imagemagick')
```

```python
# Example 1.14: NIfTI with SimpleITK
import SimpleITK as sitk

# Load image and mask
sitk_img = sitk.ReadImage("./data/image_middle_slice.nii.gz")
sitk_mask = sitk.ReadImage("./data/mask_middle_slice.nii.gz")

# Extract spacing, direction, and size
print("Size (voxel count):", sitk_img.GetSize())
print("Spacing (mm):", sitk_img.GetSpacing())
print("Direction:", sitk_img.GetDirection())

# Convert to NumPy for visualization
img_np = sitk.GetArrayFromImage(sitk_img)  # z, y, x
mask_np = sitk.GetArrayFromImage(sitk_mask).astype(bool)

# Select a slice
slice_idx = img_np.shape[0] // 2
masked_img = np.where(mask_np[slice_idx, :, :], img_np[slice_idx, :, :], np.nan)

plt.imshow(masked_img, cmap="gray")
plt.title("Example 11: Masked Slice (SimpleITK)")
plt.show()

# Exercise 2.13:
# Load a 3D NIfTI volume and mask with SimpleITK.
# - Print size, spacing, and direction
# - Select any slice and show the masked region
# - Calculate voxel volume
# - Explain differences in orientation between NiBabel and SITK

sitk_img = sitk.ReadImage("./data/phantom.nii.gz")
sitk_mask = sitk.ReadImage("./data/mask.nii.gz")

print("Size (voxel count):", sitk_img.GetSize())
print("Spacing (mm):", sitk_img.GetSpacing())
print("Direction:", sitk_img.GetDirection())

img_np = sitk.GetArrayFromImage(sitk_img)  # z, y, x
mask_np = sitk.GetArrayFromImage(sitk_mask).astype(bool)
slice_idx = img_np.shape[0] // 2
masked_img = np.where(mask_np[slice_idx, :, :], img_np[slice_idx, :, :], np.nan)

plt.imshow(masked_img, cmap="gray")
plt.title("Masked Slice (SimpleITK)")
plt.show()

voxel_dims = sitk_img.GetSpacing()
voxel_volume = np.prod(voxel_dims)
print(voxel_volume)

print(f'Diff {voxel_volume - voxel_volume_nibabel:.8f}')
```
